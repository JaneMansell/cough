Read in dataframe created previously.

```python
df_train = pd.read_csv(os.path.join(folder, 'df_train.csv'))
print(df_train.head(10))
print(df_train.shape)

df_test = pd.read_csv(os.path.join(folder, 'df_test.csv'))
print(df_test.head(10))
print(df_test.shape)
```

```python
# importing the module
import json
 
# Opening JSON file
with open(os.path.join(folder, 'rawMFCCs.json')) as json_file:
    data = json.load(json_file)

data
```

```python
df_train['MFCC1'] = np.NaN
df_train['MFCC2'] = np.NaN
df_train['MFCC3'] = np.NaN
df_train['MFCC4'] = np.NaN
df_train['MFCC5'] = np.NaN
df_train['MFCC6'] = np.NaN
df_train['MFCC7'] = np.NaN
df_train['MFCC8'] = np.NaN
df_train['MFCC9'] = np.NaN
df_train['MFCC10'] = np.NaN
df_train['MFCC11'] = np.NaN
df_train['MFCC12'] = np.NaN
df_train['MFCC13'] = np.NaN
df_train.head(5)
```

```python
df_test['MFCC1'] = np.NaN
df_test['MFCC2'] = np.NaN
df_test['MFCC3'] = np.NaN
df_test['MFCC4'] = np.NaN
df_test['MFCC5'] = np.NaN
df_test['MFCC6'] = np.NaN
df_test['MFCC7'] = np.NaN
df_test['MFCC8'] = np.NaN
df_test['MFCC9'] = np.NaN
df_test['MFCC10'] = np.NaN
df_test['MFCC11'] = np.NaN
df_test['MFCC12'] = np.NaN
df_test['MFCC13'] = np.NaN
df_test.head(5)
```

```python
def populate_MFCCs(row):
  row.MFCC1 = data[row.id]['rawMFCC'][0]
  row.MFCC2 = data[row.id]['rawMFCC'][1]
  row.MFCC3 = data[row.id]['rawMFCC'][2]
  row.MFCC4 = data[row.id]['rawMFCC'][3]
  row.MFCC5 = data[row.id]['rawMFCC'][4]
  row.MFCC6 = data[row.id]['rawMFCC'][5]
  row.MFCC7 = data[row.id]['rawMFCC'][6]
  row.MFCC8 = data[row.id]['rawMFCC'][7]
  row.MFCC9 = data[row.id]['rawMFCC'][8]
  row.MFCC10 = data[row.id]['rawMFCC'][9]
  row.MFCC11 = data[row.id]['rawMFCC'][10]
  row.MFCC12 = data[row.id]['rawMFCC'][11]
  row.MFCC13 = data[row.id]['rawMFCC'][12]
  return row

df_train_mean_MFCCs = df_train.apply(populate_MFCCs, axis = 'columns')
df_test_mean_MFCCs = df_test.apply(populate_MFCCs, axis = 'columns')

print(df_train_mean_MFCCs.head(5))
print(df_test_mean_MFCCs.head(5))  
```

```python
# save dataframes for future use
df_train_mean_MFCCs.to_csv(os.path.join(folder, 'df_train_mean_MFCCs.csv')) 
df_test_mean_MFCCs.to_csv(os.path.join(folder, 'df_test_mean_MFCCs.csv')) 
```
