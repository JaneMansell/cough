```python
import os
import pandas as pd

folder = '/content/drive/My Drive/Colab Notebooks/COVID-19'
if not os.path.exists(folder):
  os.mkdir(folder)

coughvid_folder = '/content/drive/My Drive/Colab Notebooks/coughvid'
if not os.path.exists(coughvid_folder):
  print(coughvid_folder + ' does not exist')

coswara_folder = '/content/drive/My Drive/Colab Notebooks/coswara'
if not os.path.exists(coswara_folder):
  print(coswara_folder + ' does not exist')

virufy_folder = '/content/drive/My Drive/Colab Notebooks/virufy'
if not os.path.exists(virufy_folder):
  print(virufy_folder + ' does not exist')
```

```python


df_coughvid = pd.read_csv(os.path.join(coughvid_folder, 'df_coughvid.csv'))
print(df_coughvid.shape)
df_coughvid.head()
```

```python
df_coswara = pd.read_csv(os.path.join(coswara_folder, 'df_coswara.csv'))
print(df_coswara.shape)
df_coswara.head()
```

```python
df_virufy = pd.read_csv(os.path.join(virufy_folder, 'df_virufy.csv'))
print(df_virufy.shape)
df_virufy.head()
```

Rename id column of coughvid

```python
df_coughvid.rename(columns={'uuid': 'id'}, inplace=True)
df_coughvid.head()
```

Add a source column so I know where it came from if needed.

```python
df_coughvid['source']= 'coughvid'
df_coswara['source']= 'coswara'
df_virufy['source']= 'virufy'

print(df_coughvid.head())
print(df_coswara.head())
print(df_virufy.head())
```

Add a cough_detected column to coswara and virufy. I don't have this information but I need the columns to match up when I join the dataframes.

If I need to eliminate images due to class imbalance this could be a useful column to help me identify images to remove.

```python
import numpy as np

df_coswara["cough_detected"] = np.nan
df_virufy["cough_detected"] = np.nan

print(df_coughvid.head())
print(df_coswara.head())
print(df_virufy.head())
```

Remove unwanted columns

```python
df_coughvid = df_coughvid.loc[:,['id','age','gender','status', 'audio_file', 'image_file','source', 'cough_detected']]
df_coswara = df_coswara.loc[:,['id','age','gender','status', 'audio_file', 'image_file','source','cough_detected']]
df_virufy = df_virufy.loc[:,['id','age','gender','status', 'audio_file', 'image_file','source', 'cough_detected']]

print(df_coughvid.shape)
print(df_coughvid.head())
print(df_coswara.shape)
print(df_coswara.head())
print(df_virufy.shape)
print(df_virufy.head())
```

Add path to image files.

```python
def add_path(row):
  row.image_file = os.path.join(path, row.image_file)
  return row

path = os.path.join(coughvid_folder,'coughvid_images')
df_coughvid = df_coughvid.apply(add_path, axis = 'columns')
df_coughvid.head()
```

```python
pd.set_option("display.max_colwidth", None)
print(df_coughvid.loc[:3,'image_file'])
```

```python
path = os.path.join(coswara_folder,'coswara_images')
df_coswara = df_coswara.apply(add_path, axis = 'columns')
print(df_coswara.head())
print(df_coswara.loc[:3,'image_file'])

path = os.path.join(virufy_folder,'virufy_images')
df_virufy = df_virufy.apply(add_path, axis = 'columns')
print(df_virufy.head())
print(df_virufy.loc[:3,'image_file'])
```

Make one table

```python
combined_dataset = pd.concat([df_coughvid, df_coswara, df_virufy])
print(combined_dataset.shape)
combined_dataset.head()
```

Save table

```python
combined_dataset.to_csv(os.path.join(folder, 'combined_dataset.csv'))
```

Check saved correctly

```python
combined_dataset = pd.read_csv(os.path.join(folder, 'combined_dataset.csv'), index_col=0)
print(combined_dataset.groupby('status').status.count())
print(combined_dataset.shape)
combined_dataset.head()
```

Check how many rows there are of each status.

```python
print(combined_dataset.groupby('status').status.count())
print(combined_dataset.shape)
```

Make separate dataframes for COVID-19 and healthy

```python
df_covid = combined_dataset.loc[combined_dataset.status == 'COVID-19']
df_healthy = combined_dataset.loc[combined_dataset.status == 'healthy']

print(df_covid.shape)
print(df_covid.head())
print(df_healthy.shape)
print(df_healthy.head())
```

Save for future use.

```python
df_covid.to_csv(os.path.join(folder, 'df_covid.csv'))
df_healthy.to_csv(os.path.join(folder, 'df_healthy.csv'))
```
