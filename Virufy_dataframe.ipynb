```python
import os

folder = '/content/drive/My Drive/Colab Notebooks/virufy'
if not os.path.exists(folder):
  print(folder + ' does not exist')
```

```python
import pandas as pd

df_virufy = pd.read_csv(os.path.join(folder, 'virufy_metadata.csv'))
df_virufy.head()
```

Remove unwanted columns

```python
df_virufy = df_virufy.loc[:,['id','age','gender','pcr_test_result', 'cough_path']]
print(df_virufy.shape)
df_virufy.head()
```

Change path of audio files.

```python
path = os.path.join(folder,'public_dataset')
def change_path(row):
  split = row.cough_path.split("/")
  row.cough_path = os.path.join(path, split[1])
  return row

df_virufy = df_virufy.apply(change_path, axis = 'columns')
df_virufy.head()
```

```python
pd.set_option("display.max_colwidth", None)
print(df_virufy.loc[:3,'cough_path'])
```

Rename columns

```python
df_virufy.rename(columns={'cough_path': 'audio_file', 'pcr_test_result': 'status'}, inplace=True)
df_virufy.head()
```

Check how many rows there are of each status.

```python
print(df_virufy.groupby('status').status.count())
print(df_virufy.shape)
```

I cannot use any status other than negative or positive so remove those rows.

```python
df_virufy = df_virufy.dropna(subset=['status'])
print(df_virufy.groupby('status').status.count())
print(df_virufy.shape)
df_virufy.head()
```

Other datasets use healthy and COVID-19 rather than positive and negative so I will change those.

```python
df_virufy['status'] = df_virufy['status'].replace({'positive':'COVID-19', 'negative':'healthy'})
print(df_virufy.groupby('status').status.count())
print(df_virufy.shape)
df_virufy.head()
```

Check cough files exist for all ids.

```python
filenames = df_virufy.audio_file
for filename in filenames:
  if not os.path.exists(filename):
    df_virufy.drop(df_virufy[df_virufy.audio_file == filename].index, inplace=True)

print(df_virufy.groupby('status').status.count())
print(df_virufy.shape)
df_virufy.head()
```

Save dataframe

```python
folder = '/content/drive/My Drive/Colab Notebooks/virufy'
df_virufy.to_csv(os.path.join(folder, 'df_virufy.csv'))
```
