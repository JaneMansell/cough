```python
import os

folder = '/content/drive/My Drive/Colab Notebooks/coswara'
if not os.path.exists(folder):
  print(folder + ' does not exist')
```

```python
import pandas as pd

df_coswara = pd.read_csv(os.path.join(folder, 'combined_data.csv'))
df_coswara.head()
```

Remove unwanted columns

```python
df_coswara = df_coswara.loc[:,['id','a','g','covid_status']]
print(df_coswara.shape)
df_coswara.head()
```

Create column of audio files.

```python
df_coswara['audio_file'] = df_coswara.id
df_coswara.head()
```

```python
path = os.path.join(folder,'public_dataset')
def add_path(row):
  row.audio_file = os.path.join(path, row.audio_file + '.wav')
  return row

df_coswara = df_coswara.apply(add_path, axis = 'columns')
df_coswara.head()
```

```python
pd.set_option("display.max_colwidth", None)
print(df_coswara.loc[:3,'audio_file'])
```

Check how many rows there are of each status.

```python
print(df_coswara.groupby('covid_status').covid_status.count())
```

I cannot use any status other than healthy or positive so remove those rows.

```python
df_coswara = df_coswara[df_coswara['covid_status'] != 'no_resp_illness_exposed']
df_coswara = df_coswara[df_coswara['covid_status'] != 'recovered_full']
df_coswara = df_coswara[df_coswara['covid_status'] != 'resp_illness_not_identified']
print(df_coswara.groupby('covid_status').covid_status.count())
print(df_coswara.shape)
df_coswara.head()
```

I do not need to know symptoms alongside a positive COVID-19 result so I will change all positive entries to COVID-19.

```python
df_coswara['covid_status'] = df_coswara['covid_status'].replace({'positive_asymp':'COVID-19', 'positive_mild':'COVID-19', 'positive_moderate':'COVID-19'})
print(df_coswara.groupby('covid_status').covid_status.count())
print(df_coswara.shape)
df_coswara.head()
```

Rename columns

```python
df_coswara.rename(columns={'a': 'age', 'g': 'gender', 'covid_status': 'status'}, inplace=True)
df_coswara.head()
```

Check cough files exist for all ids.

```python
filenames = df_coswara.audio_file
for filename in filenames:
  if not os.path.exists(filename):
    print(filename + ' does not exist.')
```

Save dataframe

```python
folder = '/content/drive/My Drive/Colab Notebooks/coswara'
df_coswara.to_csv(os.path.join(folder, 'df_coswara.csv'))
```
